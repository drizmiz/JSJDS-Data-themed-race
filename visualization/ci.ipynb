```python
import json
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Timeline, Map, Page
from pyecharts.globals import ThemeType

page = Page(page_title='外商投资增长效益')

with open('data/country_ce.json', 'r', encoding='utf-8') as f:
    ce_dict = json.load(f)
df = pd.read_csv("data/ci_i.csv")

tl = Timeline(init_opts=opts.InitOpts(
    theme=ThemeType.INFOGRAPHIC,
    bg_color='white',
    page_title='外商投资情况增长效应（SC）'
))

for year in range(2013, 2019+1):
    map = (
        Map()
        .add("置信区间均值",
            [[ce_dict[row['国家']], row[4]] for _, row in df[df.iloc[:, 0] == year].iterrows()],
        maptype="world",
        is_map_symbol_show=False,
        )
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            title_opts=opts.TitleOpts(title=f"{year}年外商投资情况增长效应（SC）"),
            visualmap_opts=opts.VisualMapOpts(
                max_=df[df.iloc[:, 0] == year].iloc[:, 4].max(),
                min_=df[df.iloc[:, 0] == year].iloc[:, 4].min(),
            ),
            toolbox_opts=opts.ToolboxOpts(),
        )
    )
    tl.add(map, f"{year}")
page.add(tl)

tl = Timeline(init_opts=opts.InitOpts(
    theme=ThemeType.INFOGRAPHIC,
    bg_color='white',
    page_title='外商投资情况增长效应（DiD）'
))

for year in range(2013, 2019+1):
    map = (
        Map()
        .add("置信区间均值",
            [[ce_dict[row['国家']], row[7]] for _, row in df[df.iloc[:, 0] == year].iterrows()],
        maptype="world",
        is_map_symbol_show=False,
        )
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            title_opts=opts.TitleOpts(title=f"{year}年外商投资情况增长效应（DiD）"),
            visualmap_opts=opts.VisualMapOpts(
                max_=df[df.iloc[:, 0] == year].iloc[:, 7].max(),
                min_=df[df.iloc[:, 0] == year].iloc[:, 7].min(),
            ),
            toolbox_opts=opts.ToolboxOpts(),
        )
    )
    tl.add(map, f"{year}")
page.add(tl)

page.render("out/外商投资-ci.html")
```
